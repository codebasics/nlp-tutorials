```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.pipeline
```

```python
doc = nlp("Elon few to mars yesterday. he took Tesla Roadster with him")

for token in doc:
    print(token,'|',token.pos_,'|',spacy.explain(token.pos_),'|',token.tag_,'|',spacy.explain(token.tag_))
```
